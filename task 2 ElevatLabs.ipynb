#  Titanic Dataset 

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"D:/dataset/titanic-dataset.zip")

# Display the first 5 rows
df.head()
```

#  1. Summary Statistics

```python
# Summary statistics for numeric features
df.describe()
```

# 2. histograms and boxplots for numeric features.

```python
# Plot histograms for all numeric value
df.hist(figsize=(12, 8), bins=20, color='skyblue')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.show()
```

```python
num = df.select_dtypes(include='number')
# Boxplot for all numeric value
for column in num.columns:
    plt.figure()
    sns.boxplot(y=num[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()
```

#  3. pairplot/correlation matrix for feature relationships.


```python
# Select relevant features of pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.show()
```

```python
# Correlation Matrix + Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(num.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
